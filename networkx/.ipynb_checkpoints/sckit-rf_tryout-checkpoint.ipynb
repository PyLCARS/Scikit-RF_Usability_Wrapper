```python
"""
RF Circuit Builder using NetworkX for scikit-rf

SKiDL-style interface with:
- attrs-based classes with validation
- Elements are standalone until bound to a Circuit
- Bidirectional binding (circuit knows elements, elements know circuit)
- Junction nodes for multi-terminal connections
- Bus connections for differential pairs
- Reusable subcircuits
- De-embedding support via inverted S-parameters
- Touchstone file import with automatic terminal naming
"""

from __future__ import annotations

import skrf as rf
import numpy as np
import networkx as nx
from pathlib import Path
from typing import List, Dict, Optional, Callable, ClassVar, Any, TYPE_CHECKING
from attrs import define, field, validators, Factory


# Type checking helper - checks for bus-like objects
def _is_bus(obj) -> bool:
    """Check if obj is a Bus (avoids circular import issues)."""
    return hasattr(obj, 'terminals') and hasattr(obj, '__rshift__')


# =============================================================================
# TERMINAL
# =============================================================================

@define
class Terminal:
    """
    A single port/terminal on a component.
    Supports >> operator for connections within a circuit.
    """
    component: 'Component' = field(repr=lambda c: c.name)
    port_num: int = field(validator=validators.ge(0))
    name: str = field(default=None)
    
    def __attrs_post_init__(self):
        if self.name is None:
            object.__setattr__(self, 'name', f"p{self.port_num}")
    
    @property
    def circuit(self):
        return self.component.circuit
    
    def _get_circuit_for_connection(self, other):
        """Get circuit context, ensuring both sides are compatible."""
        if isinstance(other, Terminal):
            other_circuit = other.circuit
        elif isinstance(other, (Component, Junction)):
            other_circuit = other.circuit
        elif _is_bus(other):
            other_circuit = other.circuit
        else:
            other_circuit = None
        
        if self.circuit is None and other_circuit is None:
            raise ValueError(f"Cannot connect {self} >> {other}: neither bound to a circuit")
        
        if self.circuit is not None and other_circuit is not None:
            if self.circuit is not other_circuit:
                raise ValueError(f"Cannot connect across circuits: {self} in {self.circuit}, {other} in {other_circuit}")
            return self.circuit
        
        # One is bound, one isn't - auto-bind the unbound one
        if self.circuit is not None:
            if isinstance(other, Terminal):
                other.component.bind(self.circuit)
            elif isinstance(other, Component):
                other.bind(self.circuit)
            return self.circuit
        else:
            self.component.bind(other_circuit)
            return other_circuit
    
    def __rshift__(self, other):
        """Connect this terminal to another terminal, component, junction, or bus."""
        circuit = self._get_circuit_for_connection(other)
        
        if isinstance(other, Terminal):
            circuit._connect_terminals(self, other)
            return other
        elif isinstance(other, Junction):
            circuit._connect_terminal_to_junction(self, other)
            return other
        elif isinstance(other, Component):
            circuit._connect_terminals(self, other[0])
            return other[-1] if other.nports > 1 else other[0]
        elif _is_bus(other):
            raise ValueError("Cannot connect single terminal to bus, use Bus >> Bus")
        else:
            raise TypeError(f"Cannot connect Terminal to {type(other)}")
    
    def __lshift__(self, other):
        """Reverse connection."""
        if isinstance(other, Terminal):
            other >> self
        elif isinstance(other, Component):
            other[-1] >> self
        elif isinstance(other, Junction):
            other >> self
        return self


# =============================================================================
# JUNCTION
# =============================================================================

@define
class Junction:
    """
    A named node where multiple terminals meet.
    Use for T-junctions, power splits, etc.
    """
    name: str = field(validator=validators.instance_of(str))
    circuit: 'Circuit' = field(repr=lambda c: c.name if c else None)
    _node_id: str = field(init=False)
    
    def __attrs_post_init__(self):
        object.__setattr__(self, '_node_id', f"__junction__{self.name}")
    
    def __rshift__(self, other):
        """Connect junction to terminal or component."""
        if isinstance(other, Terminal):
            self.circuit._connect_terminal_to_junction(other, self)
            return other
        elif isinstance(other, Component):
            if other.circuit is None:
                other.bind(self.circuit)
            self.circuit._connect_terminal_to_junction(other[0], self)
            return other[-1] if other.nports > 1 else other[0]
        elif isinstance(other, Junction):
            self.circuit._merge_junctions(self, other)
            return other
        else:
            raise TypeError(f"Cannot connect Junction to {type(other)}")
    
    def __lshift__(self, other):
        if isinstance(other, Terminal):
            other >> self
        elif isinstance(other, Component):
            other[-1] >> self
        return self


# =============================================================================
# BUS
# =============================================================================

@define
class _BusImpl:
    """
    Internal Bus implementation. Use Bus() factory function.
    Group of terminals for parallel/differential connections.
    """
    terminals: List[Terminal] = field(validator=validators.min_len(1))
    
    def __len__(self):
        return len(self.terminals)
    
    def __getitem__(self, idx):
        return self.terminals[idx]
    
    @property
    def circuit(self):
        """Return circuit if all terminals share one, else None."""
        circuits = set(t.circuit for t in self.terminals if t.circuit is not None)
        if len(circuits) == 1:
            return circuits.pop()
        return None
    
    def __rshift__(self, other):
        """Connect bus to bus (parallel connection)."""
        if _is_bus(other):
            if len(self) != len(other):
                raise ValueError(f"Bus width mismatch: {len(self)} vs {len(other)}")
            for t1, t2 in zip(self.terminals, other.terminals):
                t1 >> t2
            return other
        elif isinstance(other, Component):
            if len(self) > other.nports:
                raise ValueError(f"Bus width {len(self)} > component ports {other.nports}")
            for i, t in enumerate(self.terminals):
                t >> other[i]
            return other
        else:
            raise TypeError(f"Cannot connect Bus to {type(other)}")


def Bus(*terminals) -> _BusImpl:
    """
    Create a Bus (group of terminals for parallel/differential connections).
    
    Accepts:
        Bus(t1, t2, t3)      # varargs
        Bus([t1, t2, t3])    # list
    
    Example:
        Bus(dline.out_pos, dline.out_neg) >> Bus(dline2.in_pos, dline2.in_neg)
    """
    if not terminals:
        raise ValueError("Bus requires at least one terminal")
    
    # If first arg is a list/tuple, use it directly
    if len(terminals) == 1 and isinstance(terminals[0], (list, tuple)):
        terminal_list = list(terminals[0])
    else:
        terminal_list = list(terminals)
    
    if not terminal_list:
        raise ValueError("Bus requires at least one terminal")
    
    return _BusImpl(terminals=terminal_list)


# =============================================================================
# COMPONENT BASE CLASS
# =============================================================================

@define(slots=False)
class Component:
    """
    Base wrapper for scikit-rf Network objects.
    Must be bound to a Circuit before connections can be made.
    
    Set deembed=True to use inverted S-parameters (for de-embedding fixtures).
    """
    _network: rf.Network = field(repr=False, alias='network')
    deembed: bool = field(default=False, validator=validators.instance_of(bool))
    circuit: Optional['Circuit'] = field(default=None, init=False, repr=lambda c: c.name if c else None)
    _terminals: List[Terminal] = field(factory=list, init=False, repr=False)
    
    # Override in subclasses
    terminal_names: ClassVar[Optional[List[str]]] = None
    
    def __attrs_post_init__(self):
        self.name = self._network.name
        self.nports = self._network.nports
        
        # Determine terminal names
        names = self._get_terminal_names()
        
        if len(names) != self.nports:
            raise ValueError(f"terminal_names length ({len(names)}) != nports ({self.nports})")
        
        # Create terminals
        for i, tname in enumerate(names):
            term = Terminal(self, i, tname)
            self._terminals.append(term)
            setattr(self, tname, term)
    
    def _get_terminal_names(self) -> List[str]:
        """Get terminal names, checking class default, network metadata, or generating."""
        # 1. Check class-level terminal_names
        if self.terminal_names is not None:
            return self.terminal_names
        
        # 2. Check if network has port_names metadata
        if hasattr(self._network, 'port_names') and self._network.port_names:
            # Clean up port names (remove special chars, spaces)
            names = []
            for pn in self._network.port_names:
                clean = str(pn).replace(' ', '_').replace('-', '_')
                # Make valid Python identifier
                if not clean.isidentifier():
                    clean = f"p{len(names)}"
                names.append(clean)
            return names
        
        # 3. Default to p0, p1, p2...
        return [f"p{i}" for i in range(self.nports)]
    
    @property
    def network(self) -> rf.Network:
        """Return network, inverted if deembed=True."""
        if self.deembed:
            return self._network.inv
        return self._network
    
    @network.setter
    def network(self, value: rf.Network):
        self._network = value
    
    def __getitem__(self, key):
        return self._terminals[key]
    
    def bind(self, circuit: 'Circuit') -> 'Component':
        """Bind this component to a circuit."""
        if self.circuit is not None:
            if self.circuit is circuit:
                return self
            raise ValueError(f"{self.name} already bound to {self.circuit}")
        self.circuit = circuit
        circuit._add_component(self)
        return self
    
    def unbind(self) -> 'Component':
        """Remove this component from its circuit."""
        if self.circuit is not None:
            self.circuit._remove_component(self)
            self.circuit = None
        return self
    
    def bus(self, *terminal_names) -> Bus:
        """Create a bus from named terminals."""
        terminals = []
        for name in terminal_names:
            if hasattr(self, name):
                terminals.append(getattr(self, name))
            else:
                raise ValueError(f"No terminal '{name}' on {self.name}")
        return Bus(terminals)
    
    @property
    def diff_in(self) -> Bus:
        """Convenience for differential input bus (if applicable)."""
        if hasattr(self, 'in_pos') and hasattr(self, 'in_neg'):
            return Bus([self.in_pos, self.in_neg])
        raise AttributeError(f"{self.name} has no differential inputs")
    
    @property
    def diff_out(self) -> Bus:
        """Convenience for differential output bus (if applicable)."""
        if hasattr(self, 'out_pos') and hasattr(self, 'out_neg'):
            return Bus([self.out_pos, self.out_neg])
        raise AttributeError(f"{self.name} has no differential outputs")
    
    def __rshift__(self, other):
        return self._terminals[-1] >> other
    
    def __lshift__(self, other):
        return self._terminals[0] << other
    
    @classmethod
    def from_network(cls, network: rf.Network, 
                     name: Optional[str] = None,
                     terminal_names: Optional[List[str]] = None,
                     deembed: bool = False) -> 'Component':
        """
        Create a Component from an existing scikit-rf Network.
        
        Parameters
        ----------
        network : rf.Network
            The scikit-rf Network to wrap
        name : str, optional
            Override the component name (default: use network.name)
        terminal_names : list, optional
            Override terminal names (default: auto-detect from network or p0, p1, ...)
        deembed : bool
            If True, use inverted S-parameters when building circuit
        
        Returns
        -------
        Component wrapping the network
        """
        # Handle name override
        if name:
            network.name = name
        
        nports = network.nports
        
        if terminal_names:
            # Create a dynamic class with custom terminal names
            class_name = f"Component{nports}Port"
            comp_class = type(class_name, (Component,), {'terminal_names': terminal_names})
            return comp_class(network=network, deembed=deembed)
        
        # Use generic class, will auto-detect names in __attrs_post_init__
        return cls(network=network, deembed=deembed)
    
    @classmethod
    def from_touchstone(cls, filepath: str,
                        name: Optional[str] = None,
                        terminal_names: Optional[List[str]] = None,
                        deembed: bool = False) -> 'Component':
        """
        Load a component from a Touchstone file (.s1p, .s2p, etc).
        
        Parameters
        ----------
        filepath : str
            Path to Touchstone file
        name : str, optional
            Override the component name (default: filename stem)
        terminal_names : list, optional
            Override terminal names (default: auto-detect or p0, p1, ...)
        deembed : bool
            If True, use inverted S-parameters
        
        Returns
        -------
        Component with appropriate number of ports
        """
        path = Path(filepath)
        network = rf.Network(str(path))
        
        if name:
            network.name = name
        elif not network.name:
            network.name = path.stem
        
        return cls.from_network(network, terminal_names=terminal_names, deembed=deembed)


# =============================================================================
# CIRCUIT
# =============================================================================

@define(slots=False)
class Circuit:
    """
    Container for components and their connections.
    Uses NetworkX graph internally.
    """
    name: str = field(default="circuit", validator=validators.instance_of(str))
    components: Dict[str, Component] = field(factory=dict, init=False, repr=False)
    junctions: Dict[str, Junction] = field(factory=dict, init=False, repr=False)
    G: nx.Graph = field(factory=nx.Graph, init=False, repr=False)
    
    def __repr__(self):
        return f"Circuit({self.name}, {len(self.components)} components)"
    
    def add(self, *components) -> 'Circuit':
        """Bind multiple components to this circuit."""
        for comp in components:
            if isinstance(comp, SubcircuitInstance):
                comp.bind(self)
            else:
                comp.bind(self)
        return self
    
    def _add_component(self, component: Component):
        """Internal: register a component."""
        if component.name in self.components:
            raise ValueError(f"Component '{component.name}' already in circuit")
        self.components[component.name] = component
    
    def _remove_component(self, component: Component):
        """Internal: unregister a component and remove its connections."""
        if component.name not in self.components:
            return
        
        # Remove all edges involving this component's terminals
        nodes_to_remove = [(component.name, i) for i in range(component.nports)]
        for node in nodes_to_remove:
            if node in self.G:
                self.G.remove_node(node)
        
        del self.components[component.name]
    
    def remove(self, component: Component) -> 'Circuit':
        """Remove a component from the circuit."""
        component.unbind()
        return self
    
    def junction(self, name: str) -> Junction:
        """Get or create a named junction."""
        if name not in self.junctions:
            self.junctions[name] = Junction(name=name, circuit=self)
        return self.junctions[name]
    
    def _connect_terminals(self, t1: Terminal, t2: Terminal):
        """Internal: connect two terminals."""
        node1 = (t1.component.name, t1.port_num)
        node2 = (t2.component.name, t2.port_num)
        self.G.add_edge(node1, node2)
    
    def _connect_terminal_to_junction(self, terminal: Terminal, junction: Junction):
        """Internal: connect terminal to junction node."""
        term_node = (terminal.component.name, terminal.port_num)
        junc_node = junction._node_id
        self.G.add_edge(term_node, junc_node)
    
    def _merge_junctions(self, j1: Junction, j2: Junction):
        """Internal: merge two junctions."""
        self.G.add_edge(j1._node_id, j2._node_id)
    
    def build(self) -> rf.Circuit:
        """Convert to scikit-rf Circuit."""
        connections = []
        
        for component in nx.connected_components(self.G):
            connection_list = []
            for node in component:
                # Skip junction nodes
                if isinstance(node, str) and node.startswith("__junction__"):
                    continue
                name, port = node
                # Use .network property which handles deembed
                network = self.components[name].network
                connection_list.append((network, port))
            
            if connection_list:
                connections.append(connection_list)
        
        return rf.Circuit(connections)


# =============================================================================
# SUBCIRCUIT
# =============================================================================

@define
class SubcircuitDef:
    """
    Definition/template for a reusable subcircuit.
    """
    name: str = field(validator=validators.instance_of(str))
    _build_func: Optional[Callable] = field(default=None, init=False, repr=False)
    _port_names: List[str] = field(factory=list, init=False, repr=False)
    
    def define(self, ports: List[str]):
        """Decorator to define the subcircuit builder function."""
        def decorator(func):
            self._build_func = func
            self._port_names = ports
            return func
        return decorator
    
    def instantiate(self, *args, **kwargs) -> 'SubcircuitInstance':
        """Create an instance of this subcircuit."""
        if self._build_func is None:
            raise ValueError(f"SubcircuitDef '{self.name}' has no build function")
        
        port_map, components = self._build_func(*args, **kwargs)
        
        return SubcircuitInstance(
            name=self.name,
            components=components,
            port_map=port_map,
            port_names=self._port_names
        )


@define(slots=False)
class SubcircuitInstance:
    """An instantiated subcircuit with exposed terminals."""
    name: str
    components: List[Component]
    port_map: Dict[str, Terminal] = field(repr=False)
    port_names: List[str] = field(repr=False)
    circuit: Optional[Circuit] = field(default=None, init=False)
    
    def __attrs_post_init__(self):
        # Expose terminals as attributes
        for port_name in self.port_names:
            if port_name in self.port_map:
                setattr(self, port_name, self.port_map[port_name])
    
    def __repr__(self):
        ports = ", ".join(self.port_map.keys())
        bound = f" in {self.circuit.name}" if self.circuit else " (unbound)"
        return f"Subcircuit({self.name}: {ports}){bound}"
    
    def bind(self, circuit: Circuit) -> 'SubcircuitInstance':
        """Bind all internal components to circuit."""
        for comp in self.components:
            if comp.circuit is None:
                comp.bind(circuit)
        self.circuit = circuit
        return self
    
    def unbind(self) -> 'SubcircuitInstance':
        """Unbind all internal components."""
        for comp in self.components:
            comp.unbind()
        self.circuit = None
        return self


# =============================================================================
# CONCRETE COMPONENTS
# =============================================================================

@define(slots=False)
class TwoPort(Component):
    """Base for 2-port components with input/output terminals."""
    terminal_names: ClassVar[List[str]] = ['input', 'output']


@define(slots=False)
class FourPortDiff(Component):
    """Base for 4-port differential components."""
    terminal_names: ClassVar[List[str]] = ['in_pos', 'in_neg', 'out_pos', 'out_neg']


@define(slots=False)
class _Port(Component):
    """Circuit port (1 terminal). Use Port() factory function."""
    terminal_names: ClassVar[List[str]] = ['t']


def Port(frequency: rf.Frequency, name: str, z0: float = 50) -> _Port:
    """
    Create a circuit port.
    
    Parameters
    ----------
    frequency : rf.Frequency
        Frequency sweep
    name : str  
        Port name (must be unique in circuit)
    z0 : float
        Port impedance (default 50 ohms)
    
    Returns
    -------
    Port component with single terminal 't'
    """
    network = rf.Circuit.Port(frequency, name=name, z0=z0)
    return _Port(network=network)


def Isolator(frequency: rf.Frequency,
             isolation_db: float = 20,
             insertion_loss_db: float = 0.5,
             name: str = 'isolator',
             deembed: bool = False) -> TwoPort:
    """
    Create a 2-port isolator (non-reciprocal).
    
    Parameters
    ----------
    frequency : rf.Frequency
        Frequency sweep
    isolation_db : float
        Reverse isolation in dB (default 20)
    insertion_loss_db : float
        Forward insertion loss in dB (default 0.5)
    name : str
        Component name
    deembed : bool
        If True, use inverted S-parameters
    
    Returns
    -------
    TwoPort with terminals 'input' and 'output'
    """
    npoints = len(frequency)
    s = np.zeros((npoints, 2, 2), dtype=complex)
    
    s21 = 10**(-insertion_loss_db / 20)
    s[:, 1, 0] = s21
    
    s12 = 10**(-(isolation_db + insertion_loss_db) / 20)
    s[:, 0, 1] = s12
    
    s[:, 0, 0] = 0.01
    s[:, 1, 1] = 0.01
    
    network = rf.Network(frequency=frequency, s=s, z0=50, name=name)
    return TwoPort(network=network, deembed=deembed)


def Amplifier(frequency: rf.Frequency,
              gain_db: float = 10,
              reverse_isolation_db: float = 30,
              name: str = 'amplifier',
              deembed: bool = False) -> Component:
    """Create a 2-port amplifier (non-reciprocal, active)."""
    npoints = len(frequency)
    s = np.zeros((npoints, 2, 2), dtype=complex)
    
    s21 = 10**(gain_db / 20)
    s[:, 1, 0] = s21
    
    s12 = 10**(-reverse_isolation_db / 20)
    s[:, 0, 1] = s12
    
    s[:, 0, 0] = 0.1
    s[:, 1, 1] = 0.15
    
    network = rf.Network(frequency=frequency, s=s, z0=50, name=name)
    return TwoPort(network=network, deembed=deembed)


def Cable(frequency: rf.Frequency,
          length_m: float,
          loss_db_per_m: float = 0.1,
          z0_cable: float = 50,
          name: str = 'cable',
          deembed: bool = False) -> Component:
    """Create a 2-port lossy transmission line cable."""
    beta = frequency.w / rf.c
    alpha = loss_db_per_m * length_m / 8.686
    gamma = alpha + 1j * beta
    
    media = rf.media.DefinedGammaZ0(frequency=frequency, z0=z0_cable, gamma=gamma)
    network = media.line(length_m, unit='m', name=name)
    return TwoPort(network=network, deembed=deembed)


def Adapter(frequency: rf.Frequency,
            z0_in: float = 50,
            z0_out: float = 75,
            name: str = 'adapter',
            deembed: bool = False) -> Component:
    """Create a 2-port impedance adapter/transformer."""
    npoints = len(frequency)
    s = np.zeros((npoints, 2, 2), dtype=complex)
    
    gamma = (z0_out - z0_in) / (z0_out + z0_in)
    s[:, 0, 0] = gamma
    s[:, 1, 1] = -gamma
    
    t = np.sqrt(1 - gamma**2)
    s[:, 0, 1] = t
    s[:, 1, 0] = t
    
    network = rf.Network(frequency=frequency, s=s, z0=[z0_in, z0_out], name=name)
    return TwoPort(network=network, deembed=deembed)


def DifferentialLine(frequency: rf.Frequency,
                     length_m: float,
                     z0_diff: float = 100,
                     z0_comm: float = 25,
                     name: str = 'diffline',
                     deembed: bool = False) -> Component:
    """Create a 4-port differential transmission line (coupled lines)."""
    beta = frequency.w / rf.c
    npoints = len(frequency)
    theta = beta * length_m
    
    s = np.zeros((npoints, 4, 4), dtype=complex)
    
    for i, th in enumerate(theta):
        c = np.cos(th)
        s_val = 1j * np.sin(th)
        
        s[i, 2, 0] = c
        s[i, 0, 2] = c
        s[i, 3, 1] = c
        s[i, 1, 3] = c
        
        coupling = 0.1 * s_val
        s[i, 3, 0] = coupling
        s[i, 0, 3] = coupling
        s[i, 2, 1] = coupling
        s[i, 1, 2] = coupling
        
        s[i, 0, 0] = 0.05
        s[i, 1, 1] = 0.05
        s[i, 2, 2] = 0.05
        s[i, 3, 3] = 0.05
    
    network = rf.Network(frequency=frequency, s=s, z0=50, name=name)
    return FourPortDiff(network=network, deembed=deembed)


# =============================================================================
# TOUCHSTONE LOADER HELPERS
# =============================================================================

def load_touchstone(filepath: str,
                    name: Optional[str] = None,
                    terminal_names: Optional[List[str]] = None,
                    deembed: bool = False) -> Component:
    """
    Load a component from a Touchstone file.
    
    This is a convenience wrapper around Component.from_touchstone().
    
    Parameters
    ----------
    filepath : str
        Path to Touchstone file (.s1p, .s2p, .s3p, .s4p, etc)
    name : str, optional
        Override component name (default: filename)
    terminal_names : list, optional
        Override terminal names
    deembed : bool
        If True, use inverted S-parameters when building circuit
    
    Returns
    -------
    Component with appropriate terminal configuration
    """
    return Component.from_touchstone(
        filepath=filepath,
        name=name,
        terminal_names=terminal_names,
        deembed=deembed
    )
```

```python
"""
RF Circuit Builder Examples

Demonstrates building and analyzing various RF circuits:
1. 1-port: Terminated load
2. 2-port: Series attenuator
3. 3-port: Wilkinson power divider
4. 4-port: Directional coupler
5. 4-port differential: Coupled transmission lines
6. 2-port de-embedding: Fixture removal
7. Non-reciprocal: Amplifier chain with isolator
8. Quadrature hybrid: 90 coupler

Run each section in separate notebook cells.
"""

import numpy as np
import matplotlib.pyplot as plt
import skrf as rf

# Common frequency setup
freq = rf.Frequency(start=0.1, stop=4, npoints=401, unit='GHz')
z0 = 50

# %% ==========================================================================
# EXAMPLE 1: 1-Port Terminated Load
# =============================================================================
print("=" * 60)
print("EXAMPLE 1: 1-Port Terminated Load")
print("=" * 60)

def create_load(frequency, r_load, name='load'):
    """Create a 1-port resistive load."""
    npoints = len(frequency)
    gamma = (r_load - z0) / (r_load + z0)
    s = np.zeros((npoints, 1, 1), dtype=complex)
    s[:, 0, 0] = gamma
    network = rf.Network(frequency=frequency, s=s, z0=z0, name=name)
    
    @classmethod
    def _get_terminal_names(cls):
        return ['t']
    
    # Create component with single terminal
    class OnePort(Component):
        terminal_names = ['t']
    
    return OnePort(network=network)

circ1 = Circuit("terminated_load")

p1 = Port(freq, 'port1')
cable = Cable(freq, length_m=0.1, name='cable')
load_50 = create_load(freq, 50, name='load_50ohm')
load_75 = create_load(freq, 75, name='load_75ohm')
load_100 = create_load(freq, 100, name='load_100ohm')

# Test with 75 ohm load
circ1.add(p1, cable, load_75)
p1.t >> cable >> load_75.t

ntwk1 = circ1.build().network

print(f"1-port network: {ntwk1.nports} port")
print(f"S11 at 1 GHz: {20*np.log10(np.abs(ntwk1.s[100, 0, 0])):.2f} dB")

# Compare different loads
plt.figure(figsize=(10, 6))
for r_load, color in [(50, 'green'), (75, 'blue'), (100, 'red'), (25, 'orange')]:
    circ_temp = Circuit(f"load_{r_load}")
    p = Port(freq, 'p')
    cab = Cable(freq, 0.1, name='cab')
    ld = create_load(freq, r_load, f'load_{r_load}')
    circ_temp.add(p, cab, ld)
    p.t >> cab >> ld.t
    n = circ_temp.build().network
    plt.plot(freq.f/1e9, 20*np.log10(np.abs(n.s[:, 0, 0])), 
             color=color, label=f'{r_load}  load')

plt.xlabel('Frequency (GHz)')
plt.ylabel('|S11| (dB)')
plt.title('Example 1: 1-Port Terminated Load - Return Loss vs Load Impedance')
plt.legend()
plt.grid(True)
plt.ylim(-40, 0)
plt.show()

# %% ==========================================================================
# EXAMPLE 2: 2-Port Series Attenuator
# =============================================================================
print("\n" + "=" * 60)
print("EXAMPLE 2: 2-Port Series Attenuator")
print("=" * 60)

def create_attenuator(frequency, atten_db, name='atten'):
    """Create a 2-port ideal attenuator."""
    npoints = len(frequency)
    s = np.zeros((npoints, 2, 2), dtype=complex)
    
    # Transmission with attenuation
    s21 = 10**(-atten_db / 20)
    s[:, 1, 0] = s21
    s[:, 0, 1] = s21
    
    # Well matched
    s[:, 0, 0] = 0.01
    s[:, 1, 1] = 0.01
    
    network = rf.Network(frequency=frequency, s=s, z0=z0, name=name)
    return TwoPort(network=network)

circ2 = Circuit("attenuator_chain")

p1 = Port(freq, 'port1')
p2 = Port(freq, 'port2')
atten1 = create_attenuator(freq, 3, name='atten_3dB')
atten2 = create_attenuator(freq, 6, name='atten_6dB')
cable1 = Cable(freq, 0.05, name='cable1')

circ2.add(p1, p2, atten1, atten2, cable1)
p1.t >> atten1 >> cable1 >> atten2 >> p2.t

ntwk2 = circ2.build().network

print(f"2-port network: {ntwk2.nports} ports")
print(f"Total insertion loss: {-20*np.log10(np.abs(ntwk2.s[100, 1, 0])):.1f} dB")
print(f"(Expected: ~9 dB from 3+6 dB attenuators)")

plt.figure(figsize=(10, 6))
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk2.s[:, 1, 0])), 'b-', label='S21', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk2.s[:, 0, 1])), 'b--', label='S12')
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk2.s[:, 0, 0])), 'r-', label='S11')
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk2.s[:, 1, 1])), 'r--', label='S22')
plt.xlabel('Frequency (GHz)')
plt.ylabel('Magnitude (dB)')
plt.title('Example 2: 2-Port Cascaded Attenuators (3dB + 6dB)')
plt.legend()
plt.grid(True)
plt.ylim(-50, 5)
plt.show()

# %% ==========================================================================
# EXAMPLE 3: 3-Port Wilkinson Power Divider
# =============================================================================
print("\n" + "=" * 60)
print("EXAMPLE 3: 3-Port Wilkinson Power Divider")
print("=" * 60)

def create_wilkinson(frequency, z0=50, name='wilkinson'):
    """
    Create a 3-port Wilkinson power divider.
    Port 0: Input
    Port 1: Output 1 (-3dB)
    Port 2: Output 2 (-3dB)
    """
    # Design parameters
    z_branch = z0 * np.sqrt(2)  # Branch impedance
    r_isolation = 2 * z0  # Isolation resistor
    
    # Use quarter-wave at center frequency
    f_center = frequency.f[len(frequency.f)//2]
    wavelength = rf.c / f_center
    quarter_wave = wavelength / 4
    
    npoints = len(frequency)
    s = np.zeros((npoints, 3, 3), dtype=complex)
    
    # Calculate S-parameters vs frequency
    for i, f in enumerate(frequency.f):
        beta = 2 * np.pi * f / rf.c
        theta = beta * quarter_wave
        
        # Ideal Wilkinson at center frequency
        # At center freq: S11=0, S21=S31=-j/sqrt(2), S23=0
        # Away from center, performance degrades
        
        cos_t = np.cos(theta)
        sin_t = np.sin(theta)
        
        # Simplified model based on transmission line theory
        # Input match (S11)
        s[i, 0, 0] = 0.05 * (1 - np.sinc((f - f_center) / f_center))
        
        # Equal split to outputs (S21, S31)
        split = -1j * sin_t / np.sqrt(2) / (cos_t + 0.01)
        # Normalize and add frequency dependence
        split_mag = 1/np.sqrt(2) * np.exp(-0.1 * np.abs(f - f_center) / f_center)
        s[i, 1, 0] = -1j * split_mag
        s[i, 2, 0] = -1j * split_mag
        s[i, 0, 1] = s[i, 1, 0]
        s[i, 0, 2] = s[i, 2, 0]
        
        # Output match
        s[i, 1, 1] = 0.05
        s[i, 2, 2] = 0.05
        
        # Isolation between outputs (S23, S32) - should be low
        isolation = 0.05 * np.exp(-0.5 * np.abs(f - f_center) / f_center)
        s[i, 2, 1] = isolation
        s[i, 1, 2] = isolation
    
    network = rf.Network(frequency=frequency, s=s, z0=z0, name=name)
    
    class ThreePort(Component):
        terminal_names = ['input', 'out1', 'out2']
    
    return ThreePort(network=network)

circ3 = Circuit("power_divider")

p_in = Port(freq, 'p_in')
p_out1 = Port(freq, 'p_out1')
p_out2 = Port(freq, 'p_out2')
wilk = create_wilkinson(freq, name='wilkinson')

circ3.add(p_in, p_out1, p_out2, wilk)
p_in.t >> wilk.input
wilk.out1 >> p_out1.t
wilk.out2 >> p_out2.t

ntwk3 = circ3.build().network

print(f"3-port network: {ntwk3.nports} ports")
idx_2ghz = np.argmin(np.abs(freq.f - 2e9))
print(f"At 2 GHz:")
print(f"  S21 (split): {20*np.log10(np.abs(ntwk3.s[idx_2ghz, 1, 0])):.1f} dB")
print(f"  S31 (split): {20*np.log10(np.abs(ntwk3.s[idx_2ghz, 2, 0])):.1f} dB")
print(f"  S32 (isolation): {20*np.log10(np.abs(ntwk3.s[idx_2ghz, 2, 1])):.1f} dB")

plt.figure(figsize=(10, 6))
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk3.s[:, 1, 0])), 'b-', label='S21 (inputout1)', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk3.s[:, 2, 0])), 'g--', label='S31 (inputout2)', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk3.s[:, 0, 0])), 'r-', label='S11 (input match)')
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk3.s[:, 2, 1])), 'm-', label='S32 (isolation)')
plt.axhline(-3, color='gray', linestyle=':', label='-3 dB reference')
plt.xlabel('Frequency (GHz)')
plt.ylabel('Magnitude (dB)')
plt.title('Example 3: 3-Port Wilkinson Power Divider')
plt.legend()
plt.grid(True)
plt.ylim(-40, 5)
plt.show()

# %% ==========================================================================
# EXAMPLE 4: 4-Port Directional Coupler
# =============================================================================
print("\n" + "=" * 60)
print("EXAMPLE 4: 4-Port Directional Coupler")
print("=" * 60)

def create_directional_coupler(frequency, coupling_db=10, name='coupler'):
    """
    Create a 4-port directional coupler.
    Port 0: Input
    Port 1: Through
    Port 2: Coupled
    Port 3: Isolated
    """
    npoints = len(frequency)
    s = np.zeros((npoints, 4, 4), dtype=complex)
    
    coupling = 10**(-coupling_db / 20)
    through = np.sqrt(1 - coupling**2)
    
    f_center = frequency.f[len(frequency.f)//2]
    
    for i, f in enumerate(frequency.f):
        # Frequency-dependent behavior
        freq_factor = 1 - 0.1 * np.abs(f - f_center) / f_center
        
        # Through path (0 -> 1)
        s[i, 1, 0] = through * freq_factor
        s[i, 0, 1] = through * freq_factor
        
        # Coupled path (0 -> 2), 90 phase shift
        s[i, 2, 0] = -1j * coupling * freq_factor
        s[i, 0, 2] = -1j * coupling * freq_factor
        
        # Through path (3 -> 2)
        s[i, 2, 3] = through * freq_factor
        s[i, 3, 2] = through * freq_factor
        
        # Coupled path (3 -> 1)
        s[i, 1, 3] = -1j * coupling * freq_factor
        s[i, 3, 1] = -1j * coupling * freq_factor
        
        # Isolated paths (should be very small)
        s[i, 3, 0] = 0.01 * (1 - freq_factor)
        s[i, 0, 3] = 0.01 * (1 - freq_factor)
        s[i, 2, 1] = 0.01 * (1 - freq_factor)
        s[i, 1, 2] = 0.01 * (1 - freq_factor)
        
        # Return loss
        s[i, 0, 0] = 0.05
        s[i, 1, 1] = 0.05
        s[i, 2, 2] = 0.05
        s[i, 3, 3] = 0.05
    
    network = rf.Network(frequency=frequency, s=s, z0=z0, name=name)
    
    class FourPort(Component):
        terminal_names = ['input', 'through', 'coupled', 'isolated']
    
    return FourPort(network=network)

circ4 = Circuit("directional_coupler")

p_in = Port(freq, 'p_in')
p_thru = Port(freq, 'p_thru')
p_coupled = Port(freq, 'p_coupled')
p_iso = Port(freq, 'p_isolated')

coupler = create_directional_coupler(freq, coupling_db=10, name='10dB_coupler')

circ4.add(p_in, p_thru, p_coupled, p_iso, coupler)
p_in.t >> coupler.input
coupler.through >> p_thru.t
coupler.coupled >> p_coupled.t
coupler.isolated >> p_iso.t

ntwk4 = circ4.build().network

print(f"4-port network: {ntwk4.nports} ports")
idx_2ghz = np.argmin(np.abs(freq.f - 2e9))
print(f"At 2 GHz:")
print(f"  S21 (through): {20*np.log10(np.abs(ntwk4.s[idx_2ghz, 1, 0])):.1f} dB")
print(f"  S31 (coupled): {20*np.log10(np.abs(ntwk4.s[idx_2ghz, 2, 0])):.1f} dB")
print(f"  S41 (isolated): {20*np.log10(np.abs(ntwk4.s[idx_2ghz, 3, 0])):.1f} dB")
directivity = 20*np.log10(np.abs(ntwk4.s[idx_2ghz, 2, 0]) / np.abs(ntwk4.s[idx_2ghz, 3, 0]))
print(f"  Directivity: {directivity:.1f} dB")

plt.figure(figsize=(10, 6))
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk4.s[:, 1, 0])), 'b-', label='S21 (through)', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk4.s[:, 2, 0])), 'g-', label='S31 (coupled)', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk4.s[:, 3, 0])), 'r-', label='S41 (isolated)', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk4.s[:, 0, 0])), 'm--', label='S11 (return loss)')
plt.axhline(-10, color='gray', linestyle=':', label='-10 dB coupling')
plt.xlabel('Frequency (GHz)')
plt.ylabel('Magnitude (dB)')
plt.title('Example 4: 4-Port Directional Coupler (10 dB)')
plt.legend()
plt.grid(True)
plt.ylim(-50, 5)
plt.show()

# %% ==========================================================================
# EXAMPLE 5: 4-Port Differential Transmission Line
# =============================================================================
print("\n" + "=" * 60)
print("EXAMPLE 5: 4-Port Differential Transmission Line")
print("=" * 60)

circ5 = Circuit("differential_line")

p_in_pos = Port(freq, 'p_in_pos')
p_in_neg = Port(freq, 'p_in_neg')
p_out_pos = Port(freq, 'p_out_pos')
p_out_neg = Port(freq, 'p_out_neg')

# Cascaded differential lines
diff1 = DifferentialLine(freq, length_m=0.03, name='diff1')
diff2 = DifferentialLine(freq, length_m=0.03, name='diff2')

circ5.add(p_in_pos, p_in_neg, p_out_pos, p_out_neg, diff1, diff2)

# Use bus connections for differential pairs
p_in_pos.t >> diff1.in_pos
p_in_neg.t >> diff1.in_neg

# Connect diff1 output to diff2 input via bus
Bus([diff1.out_pos, diff1.out_neg]) >> Bus([diff2.in_pos, diff2.in_neg])

diff2.out_pos >> p_out_pos.t
diff2.out_neg >> p_out_neg.t

ntwk5 = circ5.build().network

print(f"4-port differential network: {ntwk5.nports} ports")
idx_2ghz = np.argmin(np.abs(freq.f - 2e9))
print(f"At 2 GHz:")
print(f"  S31 (+in  +out): {20*np.log10(np.abs(ntwk5.s[idx_2ghz, 2, 0])):.2f} dB")
print(f"  S42 (-in  -out): {20*np.log10(np.abs(ntwk5.s[idx_2ghz, 3, 1])):.2f} dB")
print(f"  S41 (+in  -out): {20*np.log10(np.abs(ntwk5.s[idx_2ghz, 3, 0])):.2f} dB (crosstalk)")

plt.figure(figsize=(10, 6))
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk5.s[:, 2, 0])), 'b-', 
         label='S31 (++) differential through', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk5.s[:, 3, 1])), 'b--', 
         label='S42 (--) differential through', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk5.s[:, 3, 0])), 'r-', 
         label='S41 (+-) crosstalk')
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk5.s[:, 2, 1])), 'r--', 
         label='S32 (-+) crosstalk')
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk5.s[:, 0, 0])), 'g-', 
         label='S11 return loss', alpha=0.7)
plt.xlabel('Frequency (GHz)')
plt.ylabel('Magnitude (dB)')
plt.title('Example 5: 4-Port Cascaded Differential Lines')
plt.legend()
plt.grid(True)
plt.ylim(-50, 5)
plt.show()

# %% ==========================================================================
# EXAMPLE 6: 2-Port De-embedding
# =============================================================================
print("\n" + "=" * 60)
print("EXAMPLE 6: 2-Port De-embedding")
print("=" * 60)

# Step 1: Create the "true" DUT
print("Creating DUT with 12 dB gain...")
dut_true = Amplifier(freq, gain_db=12, name='dut_true')

# Step 2: Create fixtures (cables that add loss and phase)
print("Creating fixtures (0.15m cables with 0.5 dB/m loss)...")

# Simulate measurement with fixtures
circ_meas = Circuit("measured_system")
p1 = Port(freq, 'p1')
p2 = Port(freq, 'p2')
fix_in = Cable(freq, length_m=0.15, loss_db_per_m=0.5, name='fixture_in')
fix_out = Cable(freq, length_m=0.15, loss_db_per_m=0.5, name='fixture_out')
dut = Amplifier(freq, gain_db=12, name='dut')

circ_meas.add(p1, p2, fix_in, fix_out, dut)
p1.t >> fix_in >> dut >> fix_out >> p2.t

measured = circ_meas.build().network

# Step 3: De-embed using inverted fixtures
print("De-embedding with inverted fixture S-parameters...")
circ_deembed = Circuit("deembedded")
p1_de = Port(freq, 'p1')
p2_de = Port(freq, 'p2')

# Same fixtures but with deembed=True (inverted S-params)
fix_in_inv = Cable(freq, length_m=0.15, loss_db_per_m=0.5, name='fix_in_inv', deembed=True)
fix_out_inv = Cable(freq, length_m=0.15, loss_db_per_m=0.5, name='fix_out_inv', deembed=True)

# Wrap measured data as a component
measured_comp = Component.from_network(measured, terminal_names=['input', 'output'])
measured_comp._network.name = 'measured'
measured_comp.name = 'measured'

circ_deembed.add(p1_de, p2_de, fix_in_inv, fix_out_inv, measured_comp)
p1_de.t >> fix_in_inv >> measured_comp >> fix_out_inv >> p2_de.t

deembedded = circ_deembed.build().network

# Compare results
idx_2ghz = np.argmin(np.abs(freq.f - 2e9))
true_gain = 12.0
meas_s21 = 20*np.log10(np.abs(measured.s[idx_2ghz, 1, 0]))
deemb_s21 = 20*np.log10(np.abs(deembedded.s[idx_2ghz, 1, 0]))

print(f"\nAt 2 GHz:")
print(f"  True DUT gain:    {true_gain:.1f} dB")
print(f"  Measured S21:     {meas_s21:.1f} dB (with fixture losses)")
print(f"  De-embedded S21:  {deemb_s21:.1f} dB (fixtures removed)")
print(f"  Error:            {np.abs(deemb_s21 - true_gain):.2f} dB")

plt.figure(figsize=(10, 6))
plt.plot(freq.f/1e9, 20*np.log10(np.abs(measured.s[:, 1, 0])), 'b-', 
         label='Measured (with fixtures)', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(deembedded.s[:, 1, 0])), 'r--', 
         label='De-embedded', linewidth=2)
plt.axhline(12, color='green', linestyle=':', label='True DUT gain (12 dB)', linewidth=2)
plt.xlabel('Frequency (GHz)')
plt.ylabel('|S21| (dB)')
plt.title('Example 6: De-embedding - Fixture Removal')
plt.legend()
plt.grid(True)
plt.show()

# Phase comparison
plt.figure(figsize=(10, 6))
plt.plot(freq.f/1e9, np.angle(measured.s[:, 1, 0], deg=True), 'b-', 
         label='Measured phase', linewidth=2)
plt.plot(freq.f/1e9, np.angle(deembedded.s[:, 1, 0], deg=True), 'r--', 
         label='De-embedded phase', linewidth=2)
plt.xlabel('Frequency (GHz)')
plt.ylabel('Phase S21 (degrees)')
plt.title('Example 6: De-embedding - Phase Correction')
plt.legend()
plt.grid(True)
plt.show()

# %% ==========================================================================
# EXAMPLE 7: Non-Reciprocal Circuit (Amplifier + Isolator Chain)
# =============================================================================
print("\n" + "=" * 60)
print("EXAMPLE 7: Non-Reciprocal Circuit")
print("=" * 60)

circ7 = Circuit("nonreciprocal")

p1 = Port(freq, 'p1')
p2 = Port(freq, 'p2')

# Build a non-reciprocal chain: isolator -> amplifier -> isolator
iso1 = Isolator(freq, isolation_db=25, insertion_loss_db=0.3, name='iso1')
amp = Amplifier(freq, gain_db=20, reverse_isolation_db=35, name='amp')
iso2 = Isolator(freq, isolation_db=25, insertion_loss_db=0.3, name='iso2')
cable1 = Cable(freq, 0.02, name='cable1')
cable2 = Cable(freq, 0.02, name='cable2')

circ7.add(p1, p2, iso1, amp, iso2, cable1, cable2)
p1.t >> iso1 >> cable1 >> amp >> cable2 >> iso2 >> p2.t

ntwk7 = circ7.build().network

idx_2ghz = np.argmin(np.abs(freq.f - 2e9))
s21_db = 20*np.log10(np.abs(ntwk7.s[idx_2ghz, 1, 0]))
s12_db = 20*np.log10(np.abs(ntwk7.s[idx_2ghz, 0, 1]))
nonrecip = s21_db - s12_db

print(f"At 2 GHz:")
print(f"  S21 (forward):  {s21_db:.1f} dB")
print(f"  S12 (reverse):  {s12_db:.1f} dB")
print(f"  Non-reciprocity (S21 - S12): {nonrecip:.1f} dB")
print(f"  (Expected: ~20 dB amp gain + 225 dB isolation + 35 dB amp reverse iso)")

plt.figure(figsize=(10, 6))
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk7.s[:, 1, 0])), 'b-', 
         label='S21 (forward)', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk7.s[:, 0, 1])), 'r--', 
         label='S12 (reverse)', linewidth=2)
plt.fill_between(freq.f/1e9, 
                 20*np.log10(np.abs(ntwk7.s[:, 1, 0])),
                 20*np.log10(np.abs(ntwk7.s[:, 0, 1])),
                 alpha=0.3, color='green', label='Non-reciprocity')
plt.xlabel('Frequency (GHz)')
plt.ylabel('Magnitude (dB)')
plt.title('Example 7: Non-Reciprocal Chain (Isolator-Amp-Isolator)')
plt.legend()
plt.grid(True)
plt.show()

# %% ==========================================================================
# EXAMPLE 8: Quadrature Hybrid (90 Coupler)
# =============================================================================
print("\n" + "=" * 60)
print("EXAMPLE 8: Quadrature Hybrid (90 Coupler)")
print("=" * 60)

# Use scikit-rf's built-in QuadratureHybrid
from skrf.media import DefinedGammaZ0

# Create media for the hybrid
line_media = DefinedGammaZ0(frequency=freq, z0=z0, gamma=1j*freq.w/rf.c)

# Create quadrature hybrid using scikit-rf's device
# The QuadratureHybrid is accessed via the media object
quad_hybrid_ntwk = line_media.quadrature_hybrid(name='quad_hybrid')

# Wrap it as our Component
class QuadHybrid(Component):
    terminal_names = ['input', 'direct', 'coupled', 'isolated']

quad_hybrid = QuadHybrid(network=quad_hybrid_ntwk)

circ8 = Circuit("quad_hybrid_circuit")

p_in = Port(freq, 'p_in')
p_direct = Port(freq, 'p_direct')
p_coupled = Port(freq, 'p_coupled')
p_iso = Port(freq, 'p_isolated')

circ8.add(p_in, p_direct, p_coupled, p_iso, quad_hybrid)
p_in.t >> quad_hybrid.input
quad_hybrid.direct >> p_direct.t
quad_hybrid.coupled >> p_coupled.t
quad_hybrid.isolated >> p_iso.t

ntwk8 = circ8.build().network

idx_2ghz = np.argmin(np.abs(freq.f - 2e9))
print(f"Quadrature Hybrid at 2 GHz:")
print(f"  S21 (direct):   {20*np.log10(np.abs(ntwk8.s[idx_2ghz, 1, 0])):.1f} dB, "
      f"phase: {np.angle(ntwk8.s[idx_2ghz, 1, 0], deg=True):.1f}")
print(f"  S31 (coupled):  {20*np.log10(np.abs(ntwk8.s[idx_2ghz, 2, 0])):.1f} dB, "
      f"phase: {np.angle(ntwk8.s[idx_2ghz, 2, 0], deg=True):.1f}")
print(f"  S41 (isolated): {20*np.log10(np.abs(ntwk8.s[idx_2ghz, 3, 0])):.1f} dB")
print(f"  Phase difference (S31 - S21): "
      f"{np.angle(ntwk8.s[idx_2ghz, 2, 0], deg=True) - np.angle(ntwk8.s[idx_2ghz, 1, 0], deg=True):.1f}")
print(f"  (Should be ~90 for quadrature hybrid)")

# Magnitude plot
plt.figure(figsize=(10, 6))
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk8.s[:, 1, 0])), 'b-', 
         label='S21 (direct)', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk8.s[:, 2, 0])), 'g-', 
         label='S31 (coupled)', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk8.s[:, 3, 0])), 'r-', 
         label='S41 (isolated)', linewidth=2)
plt.plot(freq.f/1e9, 20*np.log10(np.abs(ntwk8.s[:, 0, 0])), 'm--', 
         label='S11 (return loss)')
plt.axhline(-3, color='gray', linestyle=':', label='-3 dB reference')
plt.xlabel('Frequency (GHz)')
plt.ylabel('Magnitude (dB)')
plt.title('Example 8: Quadrature Hybrid - Magnitude Response')
plt.legend()
plt.grid(True)
plt.ylim(-50, 5)
plt.show()

# Phase plot
plt.figure(figsize=(10, 6))
phase_21 = np.unwrap(np.angle(ntwk8.s[:, 1, 0])) * 180/np.pi
phase_31 = np.unwrap(np.angle(ntwk8.s[:, 2, 0])) * 180/np.pi
phase_diff = phase_31 - phase_21

plt.subplot(2, 1, 1)
plt.plot(freq.f/1e9, phase_21, 'b-', label='S21 phase (direct)', linewidth=2)
plt.plot(freq.f/1e9, phase_31, 'g-', label='S31 phase (coupled)', linewidth=2)
plt.xlabel('Frequency (GHz)')
plt.ylabel('Phase (degrees)')
plt.title('Example 8: Quadrature Hybrid - Phase Response')
plt.legend()
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(freq.f/1e9, phase_diff, 'r-', linewidth=2)
plt.axhline(-90, color='gray', linestyle=':', label='-90 reference')
plt.xlabel('Frequency (GHz)')
plt.ylabel('Phase Difference (degrees)')
plt.title('Phase Difference (S31 - S21) - Should be ~90')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# %% ==========================================================================
# SUMMARY
# =============================================================================
print("\n" + "=" * 60)
print("EXAMPLES COMPLETE")
print("=" * 60)
print("""
Summary of demonstrated features:
1. 1-port: Terminated loads with various impedances
2. 2-port: Cascaded attenuators  
3. 3-port: Wilkinson power divider with junction
4. 4-port: Directional coupler
5. 4-port differential: Coupled lines with Bus connections
6. De-embedding: Using deembed=True for fixture removal
7. Non-reciprocal: Isolator + Amplifier chain showing S21  S12
8. Quadrature hybrid: scikit-rf built-in 90 coupler

Key API patterns:
- Component creation: Amplifier(freq, gain_db=10, name='amp')
- Circuit binding: circ.add(comp1, comp2, ...)
- Chained connections: p1.t >> cable >> amp >> p2.t
- Junctions: j = circ.junction('node'); j >> comp1; j >> comp2
- Bus (differential): Bus([t1, t2]) >> Bus([t3, t4])
- De-embedding: Cable(..., deembed=True) inverts S-params
- From touchstone: load_touchstone('file.s2p', name='dut')
""")
```
